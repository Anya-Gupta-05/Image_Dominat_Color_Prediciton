```python
def get_dominant_color(image, k=3):
    # Reshape image to (num_pixels, 3)
    data = image.reshape((-1, 3))
    data = np.float32(data)

    # Apply KMeans
    kmeans = KMeans(n_clusters=k, n_init=10)
    kmeans.fit(data)

    # Find most frequent cluster
    counts = np.bincount(kmeans.labels_)
    dominant_color = kmeans.cluster_centers_[np.argmax(counts)]
    return dominant_color.astype(int)
def get_dominant_color_name(rgb):
    channels = ['Red', 'Green', 'Blue']
    max_index = np.argmax(rgb)
    return channels[max_index]

```

```python
def rgb_to_percent(rgb):
    return [round((val / 255) * 100, 2) for val in rgb]

import matplotlib.pyplot as plt

def plot_rgb_pie(rgb):
    labels = ['Red', 'Green', 'Blue']
    percentages = rgb_to_percent(rgb)
    colors = ['red', 'green', 'blue']

    plt.pie(percentages, labels=labels, colors=colors, autopct='%1.1f%%')
    plt.title(f'Dominant RGB Composition: {rgb}')
    plt.show()

# Example
plot_rgb_pie([123, 200, 90])
```

```python


def show_color_clusters(image, k=3):
    # Reshape the image to a 2D array of pixels
    image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
    pixels = image.reshape((-1, 3))

    # Apply KMeans
    kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)
    kmeans.fit(pixels)

    # Count each cluster label
    _, counts = np.unique(kmeans.labels_, return_counts=True)

    # Sort clusters by frequency
    sorted_indices = np.argsort(-counts)
    colors = kmeans.cluster_centers_[sorted_indices]
    counts = counts[sorted_indices]

    # Normalize counts to sum to 1 for percentage
    total = np.sum(counts)
    percentages = (counts / total)

    # Create a color bar
    bar_height = 50
    bar = np.zeros((bar_height, 300, 3), dtype="uint8")

    start_x = 0
    for (percent, color) in zip(percentages, colors):
        end_x = start_x + int(percent * 300)
        cv2.rectangle(bar, (start_x, 0), (end_x, bar_height), color.astype("uint8").tolist(), -1)
        start_x = end_x

    # Plot the image and its color bar
    plt.figure(figsize=(8, 4))
    plt.subplot(1, 2, 1)
    plt.imshow(image)
    plt.title('Original Image')
    plt.axis('off')

    plt.subplot(1, 2, 2)
    plt.imshow(bar)
    plt.title('Color Clusters')
    plt.axis('off')
    plt.show()

# Example usage for 2 images:
img_paths = ['/content/images/seg_train/seg_train/buildings/1001.jpg',
             '/content/images/seg_train/seg_train/forest/10030.jpg']

for path in img_paths:
    img = cv2.imread(path)
    if img is not None:
        img = cv2.resize(img, (100, 100))
        show_color_clusters(img, k=3)
    else:
        print(f"Could not load image at {path}")
```
